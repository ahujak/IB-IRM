```python
import torch
import os
from functools import reduce

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

from domainbed.masked_model import MaskedModel, Masks, ratio, masks_and, masks_or
```

```python
# IoU across different alg's digit mask
def digit_IoU(seed, step):
    logit_digit_ls = []
    for alg_str in alg_strs:
        save_path = "ckpt/FullColoredMNIST_type{:d}_ratio{:.2f}_{:s}".format(0, 0.9, alg_str)
        mask_path = os.path.join(save_path, "mask")
        logit_digit_ls.append(torch.load(os.path.join(mask_path, 
                        "logit_digit_seed{:d}_step{:d}.pt".format(seed, step)), map_location=torch.device('cpu')))
    mask_digit_ls = [Masks({k: (v > 0) for k, v in logit.items()}) for logit in logit_digit_ls]
    and_mask = reduce(masks_and, mask_digit_ls)
    or_mask = reduce(masks_or, mask_digit_ls)
    print("area of each alg's digit:", [ratio(m) for m in mask_digit_ls])
    print("", ratio(masks_and(mask_digit_ls[0], mask_digit_ls[1])),
          ratio(masks_or(mask_digit_ls[0], mask_digit_ls[1])))
    print("and, or, IoU:", ratio(and_mask), ratio(or_mask), ratio(and_mask) / ratio(or_mask))

digit_IoU(3, 2000)
```

```python
def IoU(seed, step, alg_str, layer_str):
    save_path = "ckpt/FullColoredMNIST_type{:d}_ratio{:.2f}_{:s}".format(0, 0.9, alg_str)
    mask_path = os.path.join(save_path, "mask")
    logit_digit = torch.load(os.path.join(mask_path, "logit_digit_seed{:d}_step{:d}.pt".format(seed, step)), map_location=torch.device('cpu'))
    logit_color = torch.load(os.path.join(mask_path, "logit_color_seed{:d}_step{:d}.pt".format(seed, step)), map_location=torch.device('cpu'))
    a = logit_digit[layer_str].detach()
    b = logit_color[layer_str].detach()
    ma = a > 0
    mb = b > 0
    return torch.logical_and(ma, mb).sum().item() / torch.logical_or(ma, mb).sum().item()

layer_strs = ['0_conv1_weight', '0_conv2_weight', '0_conv3_weight', '1_weight']
alg_strs = ['ERM', 'IRM0.5', "VREx10", 'DRO1']
```

```python
dic = {"layer":[], "IoU":[], "alg":[]}
start = 0
for seed in range(1, 5):
    for alg_str in alg_strs:
        dic["layer"] += range(4)[start:]
        dic["IoU"] += [IoU(seed, 2000, alg_str, layer_strs[l]) for l in range(4)[start:]]
        dic["alg"] += [alg_str for _ in range(4)[start:]]

sns.set(style='whitegrid', context="paper", font_scale=2,rc={"lines.linewidth": 3, "figure.figsize": (4,4)})
sns.lineplot(data=pd.DataFrame(dic), x="layer", y="IoU", hue="alg", palette=my_palette)
```

```python
dic = {"step":[], "IoU":[], "alg":[]}
start = 0
for seed in range(1, 5):
    for alg_str in alg_strs:
        dic["step"] += range(21)[start:]
        dic["IoU"] += [IoU(seed, s * 100, alg_str, "0_conv1_weight") for s in range(21)[start:]]
        dic["alg"] += [alg_str for _ in range(21)[start:]]

sns.set(style='whitegrid', context="paper", font_scale=2,rc={"lines.linewidth": 3, "figure.figsize": (4,4)})
sns.lineplot(data=pd.DataFrame(dic), x="step", y="IoU", hue="alg", palette=my_palette)
```

```python
class config:
#     seed = 0
    trial_seed = 0
    type = 0
    ratio = 0.9

#     alg_str = "ERM"
    alg_str = "DRO1"
#     alg_str = "IRM0.5"
#     alg_str = "VREx10"
args = config()

save_path = "ckpt/FullColoredMNIST_type{:d}_ratio{:.2f}_{:s}".format(args.type, args.ratio, args.alg_str)
mask_path = os.path.join(save_path, "mask")

my_palette = sns.color_palette("Paired")[0:-1:2][:4]
alg_strs = ['ERM', 'IRM0.5', "VREx10", 'DRO1']
alg_names = ["ERM", "IRM", "REx", "DRO"]
```

```python
seed = 0
logit_digit = torch.load(os.path.join(mask_path, "logit_digit_seed{:d}_step{:d}.pt".format(seed, 2000)), map_location=torch.device('cpu'))
logit_color = torch.load(os.path.join(mask_path, "logit_color_seed{:d}_step{:d}.pt".format(seed, 2000)), map_location=torch.device('cpu'))

sns.set(style='darkgrid', font_scale=2,rc={"lines.linewidth": 3, "figure.figsize": (9,2)})

a = logit_digit['1_weight'].detach()
b = logit_color['1_weight'].detach()
ma = a > 0
mb = b > 0
a = torch.sigmoid(a)
b = torch.sigmoid(b)

f, axes = plt.subplots(2, 1, gridspec_kw={})
# sns.set(font_scale=.4)
sns.heatmap(a, xticklabels=0, yticklabels=False, cbar=False, ax=axes[0])
axes[0].set_ylabel("digit", fontsize=15)
sns.heatmap(b, xticklabels=0, yticklabels=False, cbar=True, cbar_ax=f.add_axes([.91, .12, .01, .75]), ax=axes[1])
axes[1].set_ylabel("color", fontsize=15)
cbar = axes[1].collections[0].colorbar
cbar.ax.tick_params(labelsize=9)
# plt.xlabel("xx")
# plt.ylabel("yy")
plt.savefig("./figures/visual/ERM_seed0_0_fc.pdf", bbox_inches='tight')
```

```python
a = logit_digit['0_conv1_weight'].detach()
b = logit_color['0_conv1_weight'].detach()
a = a.reshape(64, -1).sigmoid()
b = b.reshape(64, -1).sigmoid()

sns.set(style='darkgrid', font_scale=2,rc={"lines.linewidth": 3, "figure.figsize": (4,4)})

f, axes = plt.subplots(1, 2, gridspec_kw={})
plt.subplots_adjust(hspace=0.3, wspace=0.2)
sns.heatmap(a, xticklabels=0, yticklabels=False, cbar=False, ax=axes[0])
axes[0].set_xlabel("digit module", fontsize=15)
sns.heatmap(b, xticklabels=0, yticklabels=False, cbar=True, cbar_ax=f.add_axes([0.96, .12, .01, .75]), ax=axes[1])
axes[1].set_xlabel("color module", fontsize=15)
cbar = axes[1].collections[0].colorbar
cbar.ax.tick_params(labelsize=9)

# plt.savefig("./figures/visual/ERM_seed0_0_conv1.pdf", bbox_inches='tight')
```
