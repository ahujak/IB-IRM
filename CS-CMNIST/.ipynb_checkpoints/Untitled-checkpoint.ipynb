```python
import torch
import os

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

from domainbed.masked_model import MaskedModel, Masks, ratio, masks_and, masks_or
```

```python
class config:
#     seed = 0
    trial_seed = 0
    type = 0
    ratio = 0.9

    alg_str = "ERM"
#     alg_str = "DRO1"
#     alg_str = "IRM0.5"
#     alg_str = "VREx10"
args = config()
```

```python
save_path = "ckpt/FullColoredMNIST_type{:d}_ratio{:.2f}_{:s}".format(args.type, args.ratio, args.alg_str)
mask_path = os.path.join(save_path, "mask")

steps = range(0, 2001, 100)

step_result_ls = torch.load(os.path.join(mask_path, "step_result_ls_seed{:d}.pt".format(args.trial_seed)))
```

```python
all_results = []
alg_strs = ['ERM', 'IRM0.5', "VREx10", 'DRO1']
for seed in range(5):
    seed_dict = {}
    for alg_str in alg_strs:
        alg_ls = torch.load(os.path.join("./ckpt/FullColoredMNIST_type{:d}_ratio{:.2f}_{:s}/mask".format(args.type, args.ratio, alg_str), "step_result_ls_seed{:d}.pt".format(seed)))
        seed_dict[alg_str] = alg_ls
    all_results.append(seed_dict)
# erm_ls = torch.load(os.path.join("./ckpt/FullColoredMNIST_type{:d}_ratio{:.2f}_{:s}/mask".format(args.type, args.ratio, "ERM"), "step_result_ls_seed{:d}.pt".format(args.trial_seed)))
# dro_ls = torch.load(os.path.join("./ckpt/FullColoredMNIST_type{:d}_ratio{:.2f}_{:s}/mask".format(args.type, args.ratio, "DRO1"), "step_result_ls_seed{:d}.pt".format(args.trial_seed)))
```

```python
dic = {"step":[], "seed":[], "acc":[], "type":[], "mask":[], "alg":[]}
alg_strs = ["ERM", "IRM0.5", "VREx10", "DRO1"]
alg_names = ["ERM", "IRM", "REx", "DRO"]
type_strs = ["digit", "color"]  # report what kind of accuracy
start = 0

for alg_idx, alg_str in enumerate(alg_strs):
    save_path = "ckpt/FullColoredMNIST_type{:d}_ratio{:.2f}_{:s}".format(0, 0.9, alg_str)
    for seed in range(5):
        result_ls = torch.load(os.path.join(save_path, "mask", "step_result_ls_seed{:d}.pt".format(seed)))
        for type_str in type_strs:
            mask_strs = ["digit", "universal"] if type_str == 'digit' else ["color", "universal"]
            for idx, mask_str in enumerate(mask_strs):
                dic['step'] += range(21)[start:]
                dic["seed"] += [seed for _ in range(21)][start:]
                dic["acc"] += [a[mask_str]["results" if type_str == "digit" else "results_color"]['env0_out_acc'] for a in result_ls][start:]
#                 dic["acc"] += [a[mask_str]["results"]['env0_out_acc'] for a in result_ls][start:]
                dic["type"] += [type_str for _ in range(21)][start:]
                dic["mask"] += [mask_str for _ in range(21)][start:]
                dic["alg"] += [alg_names[alg_idx] for _ in range(21)][start:]
```

```python
sns.set(style='darkgrid', font_scale=2,rc={"lines.linewidth": 2, "figure.figsize": (2,3)})
sns.relplot(data=pd.DataFrame(dic), x="step", y="acc", col="mask", row="type", hue="alg", kind="line")
# f, ax = plt.subplots()
# sns.lineplot(x="step", y="acc", data=pd.DataFrame(dic), hue="type")
```
